```python
import pandas as pd
```

```python
taxonomy_excel = pd.read_excel("/home/rszczygielski/bioinf/magisterka/geneBank/results/main_mitochondrion_no_location.xlsx")
taxonomy_excel1 = pd.read_excel("/home/rszczygielski/bioinf/magisterka/geneBank/results/main_mitochondrion_based_on_location.xlsx")
```

```python
joined_data = taxonomy_excel.merge(taxonomy_excel1,
                            indicator=True,
                            how="outer")

discrepancies_df = joined_data.loc[lambda x: x["_merge"] != "both"]
discrepancies_df['APPROACH'] = discrepancies_df['_merge'].apply(lambda x: 'NoLocation' if x == 'left_only' else "WithLocation")
discrepancies_df = discrepancies_df.drop('_merge', axis=1)
discrepancies_df = discrepancies_df.sort_values(["ACCESSION", "APPROACH"])
discrepancies_df = discrepancies_df.set_index("ACCESSION")
display(discrepancies_df.columns)
display(discrepancies_df)
```

```python
only_accession = discrepancies_df.drop(['TAXONOMY', 'PREVIOUS_GENE_NAME', 'PREVIOUS_GENE_LOCATION',
       'PREVIOUS_GENE_PRODUCT_NAME', 'CONTROL_REGION_NAME',
       'CONTROL_REGION_LOCATION', 'NEXT_GENE_NAME', 'NEXT_GENE_LOCATION',
       'NEXT_GENE_PRODUCT_NAME'], axis=1)
display(only_accession)
```

```python
discrepancies_df.to_excel("/home/rszczygielski/bioinf/magisterka/geneBank/data_analysis/comparison.xlsx")
only_accession.to_excel("/home/rszczygielski/bioinf/magisterka/geneBank/data_analysis/comparison_short.xlsx")
```

```python
mammals_tasxonomy_string = "RodentiaChiropteraSoricomorphaPrimatesCarnivoraArtiodactylaDiprotodontiaLagomorphaDidelphimorphiaCetaceaDasyuromorphiaAfrosoricidaErinaceomorphaCingulataPeramelemorphiaScandentiaPerissodactylaMacroscelideaPilosaMonotremataSireniaProboscidea"

def get_taxonomy_list(main_sting):
    new_str = ""
    for elem in main_sting:
        if elem.isupper():
            new_str += f" {elem}"
        else:
            new_str += elem
    return new_str.split()

mammals_taxonomy_list = get_taxonomy_list(mammals_tasxonomy_string)

mammals_df = pd.DataFrame({"Mammals": mammals_taxonomy_list})
display(mammals_df)
```
